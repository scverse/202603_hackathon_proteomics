## Aggregation demo

We provide the option to aggregate one modality into another. So for example, if given the precursor layer and a relationship between precursors and proteins (and potentially proteins to genes), you can aggregate (sum, mean, median,...) the precursor values to protein values. Often, these aggregated values are already provided by the DB search output, but there might be instances were you want to have a custom aggregation. Aggregation across multiple levels is also allowed. A possible future direction is to also pass a custom aggregation function to the aggregation.


Technically, the mudata needs to already have an "empty" anndata for the desired, aggregated level, and a "feature_mapping" varp (hence, the empty anndata: to have matching dimensions). By default, the resulting aggregation is returned (the ad.X) and the original mudata remains untouched. The aggregation function exposes several functionalities like overwriting the existing mudata to change the default behaviour. For more info, check the docstring of the function.
